# Exploratory Data Analysis


### Gender

```python
# plotting a bar chart for gender and its count
ax = sns.countplot(x = 'Gender' , data = df, hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#plotting bar chart for gender vs total amount
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender', y='Amount', hue='Gender', data = sales_gen)
sns.set(rc={'figure.figsize':(15,7)})
```

## Age

```python
#plotting bar chart for age 
ax = sns.countplot(x = 'Age Group', data=df, hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'],as_index=False).sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group', y = 'Amount', hue='Age Group', data = sales_age)
```

```python
df.columns
```

```python
ax = sns.countplot(x='Marital_Status', data=df,hue='Marital_Status')
sns.set(rc={'figure.figsize':(25,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_m = df.groupby(['Marital_Status', 'Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status', y = 'Amount', hue='Gender', data = sales_m)
```

## State

```python
ax = sns.countplot(x='State', data=df,hue='State')
sns.set(rc={'figure.figsize':(25,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y = 'Orders', hue='State', data = sales_state)
```

## Occupation

```python
ax = sns.countplot(x='Occupation', data=df,hue='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_oc = df.groupby(['Occupation'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.barplot(x='Occupation', y = 'Orders', hue='Occupation', data = sales_oc)
```

```python
ax = sns.countplot(x='Product_Category', data=df,hue='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_category = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(x='Product_Category', y='Amount', hue='Product_Category', data=Sales_category)
```

```python
sales_category_orders = df.groupby(['Product_Category'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.barplot(x='Product_Category', y='Orders', hue='Product_Category', data=sales_category_orders)
```

```python
fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_Category')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

## Marital Status

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```
